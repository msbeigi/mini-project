```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
def extract_posts(url, print_content):

  response = requests.get(url)
  content = response.text
  soup=BeautifulSoup(content,'html.parser')
  if(print_content):
    print(soup.prettify())

  tags = soup.find_all('blockquote', {'class': 'postMessage'})#.get_text()

  list_content=[]
  dict_content={}
  for c,tag in enumerate(tags):
    if tag.text.find("(USER WAS BANNED FOR THIS POST)" or "Remember: All request threads belong on ") != -1:
      break
    elif (len(tag.text)>0):
      dict_content = {"post": tag.text,"url":url}
      # print('\r\n/////dict',c,dict_content['post'])
      list_content.append(dict_content)

  df_content=pd.DataFrame(list_content)
  return df_content
```

```python
list_urls=[]
df_all_posts_other=pd.DataFrame();
```

```python

url = "https://www.4chan.org/" 
html = requests.get(url).content.decode('utf-8')
links = re.findall('"((http|ftp)s?://.*?)"', html) 
for link in links:
    list_urls.append(link[0]) # Prints the full URL of each link extracted
```

```python
for url in list_urls:
  df_posts =extract_posts(url, print_content=False)
  df_all_posts_other=pd.concat([df_all_posts_other,df_posts],axis=0)
df_all_posts_other.reset_index()
# df_all_posts_other
df_all_posts_other.to_csv('all_posts.csv')
```
